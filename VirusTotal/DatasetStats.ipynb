#After Collecting the information from VirusTotal API and save them in a csv file it is very usefull to make some diagrams.In the code above i am choosing to make a diagram based on the malware category.Although easily modifications can be made to choose any other information.First column is the code with comments so it can be easily changeable and at the second original code.

```python
import pandas as pd
import os

# Define the path to the CSV file
file_path = ''

# Check if the CSV file exists
if not os.path.exists(file_path):
    print("CSV file not found.")
    exit()  

# Load the CSV file
df = pd.read_csv(file_path)

# Handle null values by replacing them with the string 'null', Very common only in malware names if the classifier didnt know the family name
df['Kaspersky_result'] = df['Kaspersky_result'].fillna('null')

# Preprocess the column you want to remove unwanted substrings
df['your column'] = df['your column'].str.replace(r'remove|this|and|that|:', '', regex=True)

# Generate value counts including the 'null' category
name_counts = df['your column'].value_counts(dropna=False)

# Calculate the total count
total_count = name_counts.sum()

# Calculate the threshold for the "other" category (1% of the total count)
threshold = 0.01 * total_count

# Filter values that represent less than 1%
filtered_counts = name_counts[name_counts >= threshold].copy()

# Sum the counts of values that represent less than 1%
other_count = name_counts[name_counts < threshold].sum()

# Add the 'other' category if applicable
if other_count > 0:
    filtered_counts['other'] = other_count

# Optional: Sort the counts in descending order
filtered_counts = filtered_counts.sort_values(ascending=False)

# Create a DataFrame for better readability
summary_table = pd.DataFrame({
    'Category': filtered_counts.index,
    'Count': filtered_counts.values,
    'Percentage': (filtered_counts.values / total_count) * 100
})

# Reset index for a clean table
summary_table.reset_index(drop=True, inplace=True)

# Round the percentage to two decimal places
summary_table['Percentage'] = summary_table['Percentage'].round(2)

# Display the table
print("Distribution of Kaspersky Results:")
print(summary_table.to_string(index=False))

# Optional: Save the table to a CSV file
output_csv_path = '/content/drive/MyDrive/Διπλωματική-ΑΣ/Code/Stats/Kaspersky_Result_Distribution.csv'
summary_table.to_csv(output_csv_path, index=False)
print(f"\nSummary table saved to {output_csv_path}")

# Optional: Display the table in a Jupyter Notebook
# from IPython.display import display
# display(summary_table)

# Optional: Create and save a bar chart for visualization
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
plt.bar(summary_table['Category'], summary_table['Count'], color='skyblue')
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Distribution of whatever')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('/save/here/stats.png')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import os


file_path = '/Stats/VirusTotal Stats/exe_samples_346_2019.csv'



df = pd.read_csv(file_path)


df['Kaspersky_result'] = df['Kaspersky_result'].fillna('null')


df['Kaspersky_result'] = df['Kaspersky_result'].str.replace(r'Win64|Win32|HEUR|not-a-virus|:', '', regex=True)


name_counts = df['Kaspersky_result'].value_counts(dropna=False)


total_count = name_counts.sum()


threshold = 0.01 * total_count


filtered_counts = name_counts[name_counts >= threshold].copy()


other_count = name_counts[name_counts < threshold].sum()


if other_count > 0:
    filtered_counts['other'] = other_count


filtered_counts = filtered_counts.sort_values(ascending=False)


summary_table = pd.DataFrame({
    'Category': filtered_counts.index,
    'Count': filtered_counts.values,
    'Percentage': (filtered_counts.values / total_count) * 100
})


summary_table.reset_index(drop=True, inplace=True)

# Round the percentage to two decimal places
summary_table['Percentage'] = summary_table['Percentage'].round(2)


print("Distribution of Kaspersky Results:")
print(summary_table.to_string(index=False))


output_csv_path = '/Stats/Kaspersky_Result_Distribution.csv'
summary_table.to_csv(output_csv_path, index=False)
print(f"\nSummary table saved to {output_csv_path}")



plt.figure(figsize=(12, 8))
plt.bar(summary_table['Category'], summary_table['Count'], color='skyblue')
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Distribution of Kaspersky Results')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig(/Code/Stats/bar_chart_346.png')
plt.show()
```
